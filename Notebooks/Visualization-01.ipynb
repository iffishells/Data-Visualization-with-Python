# Mean ,Variance , Standard Variance
* **Mean**
    * Mean is failed to detect the outlier ,one wrong input can make the wrog the output
* **Variance**
    * Variance actually tell us the spread of the data from the mean
    * if we know the mean and variance then we can imagine of the data where they are laying
* **Standard Variance** 
    * std-var is nothing this just the square of the variance to remove the effect of the negative sign of the distance from each point from the mean of the whole data
* **Median**
    * By median you crack the mean problem what is mean problem we can go wrong if the outlier is under obervation>
    * Median hack the problem by sort it first and pick the middle value then we can observe the how much data is my left side and how much data is on my right side.
    * But there is also problem for example List = [1.1,1.2,1.3,1.5,10,20,30,40,50] then the mean of this problem is we got medain of this problem is 1.5 . Do you think this is right interpretation ?
    * No this is wrong interpretation . The point is When we have more then 50 percent data is outlier then Median would fail.
* **Percentile**
    * Percentile is nothing it is just tell us the percentage of the data. 25 percentile is the 25 percentage of the data is liying in thpose Range.
       * **Example**
           * order dlivery list = [1,1.5,2,3,4,5,6]
           * if we check th the 90 percentile is mean of the 90 percentage of the people are getting in five days



```python
petalLenght_mean = np.mean(iris.PetalLengthCm)
print(petalLenght_mean)
petal_width_mean= np.mean(iris.PetalWidthCm)
print(petal_width_mean)
```

# Boxplots 
**What is the boxplot and intutions?**  
   * Boxplot is nothing it is just representation of percentile with 25th and 50th and 75th and 100th of data.  
   * upper and lower line also called wisker . Wisker represent the max and min value and but generally have differnent methods of wisker.
    * for example whisker is defined as 2 * IQR
        * **What is IQR**?
            * IQR is datal between 75 percentile and 25 percentile.

```python
sns.boxplot(x = 'Species' ,y = 'PetalLengthCm' ,data=iris)
```
