```python
# Bar graphs are especially effective when you have numerical data that splits nicely
# into different categories, so you can quickly see trends within your data
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

N = 7

innerspot = [142.6, 125.3, 62.0, 81.0, 145.6, 319.4, 178.1]

ind =  np.arange(N) # the x location of the group

width =0.35  # the idfth for the bars



fig,ax  = plt.subplots(figsize=(15,20))

inner = ax.bar(ind ,innerspot , width=width , color = '#ffad00')

nominespot = [109.4, 1.8, 60.7, 44.6, 116.9, 262.5, 102.0]
nominees = ax.bar(ind+width, nominespot,width=width , color="blue")


# add some text fot the labels ,titile and axxes tricks
ax.set_xticks(ind+width)
ax.set_xticklabels(["best picture" , 'Director', 'best Actor' , 'Editing' , 'Visual Effect','Chinametography'])

ax.legend((inner[0] , nominees[0]),("Acadmy Aard" , "Acadmy aard Nominess"))


def autolabels(rects):
    
    # attach for some text labels
    
    for rect in rects:
        
        height = rect.get_height()
#         print(height)
        hcap = "$"+str(height)+"M"
#         print(rect.get_x())
        ax.text(rect.get_x()+rect.get_width()/2,height ,hcap ,
                ha='center',
                va='bottom',
                rotation='vertical'
               )
                # plt.show()
autolabels(inner)
autolabels(nominees)
plt.show()
```
