```python
import pandas as pd
import matplotlib.pyplot as plt

## in this way dataframe setted as index as data

climate_change = pd.read_csv("Data/climate_change.csv",
                  parse_dates=["date"],
                  index_col="date")

climate_change
```

```python
##plotting two time-series together

fix , ax = plt.subplots()

ax.plot(climate_change.index ,climate_change["co2"])
ax.plot(climate_change.index ,climate_change["relative_temp"])
ax.set_xlabel("Time")
ax.set_ylabel("CO2(ppm) /Relative temp")
plt.show()

## not looking good
```

```python
## using twins

fix , ax = plt.subplots()
ax.plot(climate_change.index ,climate_change["co2"])
ax.set_xlabel("Time")
ax.set_ylabel("CO2(ppm) /Relative temp")

ax2 = ax.twinx()
ax2.plot(climate_change.index ,climate_change["relative_temp"])
ax2.set_ylabel("Relative Temperature (Celsius)")
plt.show()
```

```python
## sperated by colors
fig, ax = plt.subplots()
ax.plot(climate_change.index, climate_change["co2"], color='blue')
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)', color='blue')
ax2 = ax.twinx()
ax2.plot(climate_change.index, climate_change["relative_temp"],
color='red')
ax2.set_ylabel('Relative temperature (Celsius)', color='red')
plt.show()
```

```python
## Coloring the ticks
fig, ax = plt.subplots()
ax.plot(climate_change.index, climate_change["co2"],
color='blue')
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)', color='blue')
ax.tick_params('y', colors='blue')
ax2 = ax.twinx()
ax2.plot(climate_change.index,
climate_change["relative_temp"],
color='red')
ax2.set_ylabel('Relative temperature (Celsius)',
color='red')
ax2.tick_params('y', colors='red')
plt.show()
```

```python
## A function that plots time-series

def plot_timeseries(axes,x,y,color,xlabel,ylabel):
    axes.plot(x, y, color=color)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel, color=color)
    axes.tick_params('y', colors=color)
    
```

```python
## using our function
fig, ax = plt.subplots()
plot_timeseries(ax, climate_change.index, climate_change['co2'],
'blue', 'Time', 'CO2 (ppm)')
ax2 = ax.twinx()
plot_timeseries(ax, climate_change.index,
climate_change['relative_temp'],
'red', 'Time', 'Relative temperature (Celsius)')
plt.show()
```
